# modeling_v5 — Lot 누수 검증 (멘토 피드백)

**질문**: v5의 RMSE ~60이 비정상적으로 높은 것 아닌가? valid로도 검증하라.

**진단**: Lot ID를 피처로 쓰진 않지만, CV를 **wafer(C64) 단위**로 나눠서 문제가 있음.
- 한 Lot당 wafer 평균 9.8장, **Lot 내부 C65 std 26.5** (전체 262) → Lot만 알면 R²≈0.99.
- GroupKFold(C64)에서 검증 wafer의 **99.9%가 train에 같은 Lot 형제 보유** → 간접 Lot 누수.
- valid/test도 train과 Lot 99.9~100% 공유 → valid 검증도 '신규 Lot' 테스트가 아님.

**이 노트북**: 동일한 v5 피처로 **GroupKFold(C64) vs GroupKFold(C20=Lot)** OOF RMSE를 비교.
두 값의 차이 = Lot 누수의 크기 = 낙관 편향의 크기. **GroupKFold(C20) 쪽이 신규 Lot 실제 성능에 가까움.**

> Lot(C20)은 검증 전용. 최종 제출 모델에는 여전히 미사용(일반화 원칙 유지).

## Cell 1 — imports & 로드

```python
import numpy as np, pandas as pd, warnings
from pathlib import Path
from sklearn.model_selection import GroupKFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
warnings.filterwarnings("ignore")
DATA_DIR=Path("../문제1(하)")
WF_ID,LOT_ID,TARGET,N_FOLDS="C64","C20","C65",5
train=pd.read_csv(DATA_DIR/"train_data.csv")
for c in ("C64","C20","C23","C6","C7"):
    train[c]=train[c].astype("object")
print("train", train.shape)
```

## Cell 2 — v5 피처 빌더 (원본 동일) + Lot 라벨 보존

```python
DROP_IDS=["C34","C35","C38"]; DROP_LOT=["C20","C21","C22"]; DROP_CONST=["C14","C24","C30"]
DROP_EXCL=["C26","C28","C29","C37"]; DROP_ALLNA=["C2","C13","C43","C47","C53","C55"]
DROP_TIME=["C10","C39","C40","C41"]; DROP_DUP=["C36"]
DROP_ALL=DROP_IDS+DROP_LOT+DROP_CONST+DROP_EXCL+DROP_ALLNA+DROP_TIME+DROP_DUP
CAT_COLS=["C6","C7"]; TE_COL="C23"

def build_rows(df):
    df=df.copy()
    excl=set([WF_ID,TARGET]+DROP_ALL+CAT_COLS+[TE_COL])
    sensors=[c for c in df.select_dtypes(include=[np.number]).columns if c not in excl]
    g=df.groupby(WF_ID)
    ctx=g[sensors].agg(["mean","std","min","max"]); ctx.columns=[f"{c}_wf_{s}" for c,s in ctx.columns]
    ctx["wf_nrows"]=g.size(); df2=df.merge(ctx,on=WF_ID,how="left")
    df2["row_pos"]=g.cumcount()
    feat_cols=sensors+list(ctx.columns)+["row_pos"]+CAT_COLS
    for c in CAT_COLS: df2[c]=df2[c].astype("category")
    out=df2[[WF_ID]+feat_cols].copy(); out[TE_COL]=df2[TE_COL].values
    out[TARGET]=df2[TARGET].values
    out[LOT_ID]=df[LOT_ID].values          # Lot 라벨(그룹키 후보, 피처 아님)
    return out, feat_cols

rows,FEATS=build_rows(train)
def te_fit(frame,col,target,m=20):
    prior=frame[target].mean(); agg=frame.groupby(col)[target].agg(["mean","count"])
    return (agg["mean"]*agg["count"]+prior*m)/(agg["count"]+m), prior
print("row 피처:", len(FEATS)+1)
```

## Cell 3 — 공통 학습 함수 (그룹키만 바꿔 재사용)

```python
PARAMS=dict(objective="regression",metric="rmse",boosting_type="gbdt",
    learning_rate=0.03,num_leaves=127,max_depth=-1,min_child_samples=50,
    subsample=0.8,subsample_freq=1,colsample_bytree=0.7,reg_alpha=0.5,reg_lambda=1.0,
    n_estimators=4000,random_state=42,n_jobs=-1,verbose=-1)
USE=FEATS+[TE_COL+"_te"]
y=rows[TARGET].values

def run_cv(group_col):
    """group_col 기준 GroupKFold로 OOF 예측 → WF 단위 RMSE 반환"""
    groups=rows[group_col].astype("category").cat.codes.values
    gkf=GroupKFold(N_FOLDS); oof=np.zeros(len(rows))
    for k,(tri,vai) in enumerate(gkf.split(rows,y,groups)):
        tf=rows.iloc[tri].copy(); vf=rows.iloc[vai].copy()
        enc,prior=te_fit(tf,TE_COL,TARGET)
        tf[TE_COL+"_te"]=tf[TE_COL].map(enc).fillna(prior); vf[TE_COL+"_te"]=vf[TE_COL].map(enc).fillna(prior)
        m=lgb.LGBMRegressor(**PARAMS)
        m.fit(tf[USE],tf[TARGET],eval_set=[(vf[USE],vf[TARGET])],categorical_feature=CAT_COLS,
              callbacks=[lgb.early_stopping(100,verbose=False),lgb.log_evaluation(0)])
        oof[vai]=m.predict(vf[USE])
        print(f"  [{group_col}] fold{k+1} done")
    # WF 단위 집계 RMSE
    d=pd.DataFrame({"wf":rows[WF_ID].to_numpy(),"oof":oof,"y":y})
    wf=d.groupby("wf").agg(oof=("oof","mean"),y=("y","first"))
    return float(np.sqrt(mean_squared_error(wf["y"],wf["oof"])))
```

## Cell 4 — 두 CV 비교 실행 (핵심)

```python
print("=== GroupKFold(C64=wafer): 같은 Lot 형제가 train에 있음 (현재 v5 방식) ===")
rmse_wafer=run_cv(WF_ID)
print("\n=== GroupKFold(C20=Lot): 통째로 새로운 Lot 예측 (신규 Lot 일반화) ===")
rmse_lot=run_cv(LOT_ID)

print("\n"+"="*55)
print(f"wafer 단위 CV  OOF RMSE : {rmse_wafer:.2f}  (낙관적, 간접 Lot 누수 포함)")
print(f"Lot   단위 CV  OOF RMSE : {rmse_lot:.2f}  (신규 Lot 실제 성능 추정)")
print(f"Lot 누수로 인한 낙관 편향: {rmse_lot-rmse_wafer:.2f} pt")
print("="*55)
```

## Cell 5 — 해석

- **두 값이 비슷하면** (예: 62 vs 64): Lot 누수 영향이 작고, v5의 ~60은 신규 Lot에도 대체로 유효. 멘토 우려는 기우.
- **Lot 단위가 크게 나쁘면** (예: 62 vs 85): 현재 성능은 같은 Lot 형제 덕분에 부풀려진 것. 신규 Lot 실제 성능은 그만큼 낮음 → 리더보드(같은 Lot)와 실무(신규 Lot) 성능을 **분리 보고**해야 함.

어느 쪽이든, **최종 제출 모델의 CV는 GroupKFold(C20=Lot)로 하는 것이 실무형 일반화 평가에 정직**합니다. (Lot은 그룹키로만 사용, 피처로는 계속 미사용.)
